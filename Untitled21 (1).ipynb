```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
croppredictorknn_df = pd.read_csv(r"C:\Users\User\Documents\croppredictorknn.csv")
croppredictorknn_df
```

```python


# Group by Crop_Type and calculate average Soil_Moisture
moisture_per_crop = croppredictorknn_df.groupby('Crop_Type')['Soil_Moisture'].mean().sort_values()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(moisture_per_crop.index, moisture_per_crop.values, color='green', marker='o', linestyle='-')
plt.title('Average Soil Moisture by Crop Type', fontsize=16)
plt.xlabel('Crop Type', fontsize=12)
plt.ylabel('Average Soil Moisture', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Set up the figure
plt.figure(figsize=(12, 6))

# Scatter plot: Crop_Type on x-axis (categorical), Soil_Moisture on y-axis
plt.scatter(croppredictorknn_df['Crop_Type'], croppredictorknn_df['Soil_Moisture'], color='blue', alpha=0.7)

# Add titles and labels
plt.title('Soil Moisture Distribution by Crop Type', fontsize=16)
plt.xlabel('Crop Type', fontsize=12)
plt.ylabel('Soil Moisture', fontsize=12)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show grid
plt.grid(True, linestyle='--', alpha=0.5)

# Tight layout to avoid label cutoff
plt.tight_layout()

# Show plot
plt.show()
```

```python
import seaborn as sns


plt.figure(figsize=(12, 6))
sns.histplot(data=croppredictorknn_df, x='Soil_Moisture', hue='Crop_Type', kde=True, palette='Set2', multiple='stack')
plt.title('Histogram of Soil Moisture by Crop Type', fontsize=16)
plt.xlabel('Soil Moisture', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data=croppredictorknn_df, x='Crop_Type', y='Soil_Moisture', palette='Set3')
plt.title('Boxplot of Soil Moisture by Crop Type', fontsize=16)
plt.xlabel('Crop Type', fontsize=12)
plt.ylabel('Soil Moisture', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
